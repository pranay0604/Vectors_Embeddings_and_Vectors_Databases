```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
v1 = np.array([[1,2]])
v2 = np.array([[2,4]])
v3 = np.array([[-1,-2]])
```

```python
# v1 and v3 are opposite --> Low

# v1 and v2 are similar --> High
```

```python
cosine_similarity(v1,v2)
```

```python
cosine_similarity(v1,v3)
```

```python
cosine_similarity(v3,v2)
```
